# Working With Dates

The dates can be used to analyse data.
Lets display the dates

```python
covid_df.date
```

the dtype is object.
Panda can not recognise it as date.
To make it recognise we will have to change the dtype from object to <code>datetime</code> using <code>pd.to_datetime</code>

```python
covid_df["date"]=pd.to_datetime(covid_df.date)
covid_df.date
```

```python
covid_df["year"]=pd.DatetimeIndex(covid_df.date).year
covid_df["month"]=pd.DatetimeIndex(covid_df.date).month
covid_df["day"]=pd.DatetimeIndex(covid_df.date).day
covid_df["weekdays"]=pd.DatetimeIndex(covid_df.date).weekday
covid_df
```

Lets aggregate the data for the month of the may

```python
covid_df_may=covid_df[covid_df.month==5]
covid_df_may_metrices=covid_df_may[["new_tests","new_deaths","new_cases"]]
covid_df_totals=covid_df_may_metrices.sum()
covid_df_totals
```

we could also do it in this way.

```python
covid_df_totals_2=covid_df[covid_df.month==5][["new_tests","new_deaths","new_cases"]].sum()
```

```python
covid_df_totals==covid_df_totals_2
```

Lets check if number of cases on sundays were greater than the mean.

```python
#overall avg
mean_new_cases=covid_df.new_cases.mean()
```

```python
covid_df_sundays=covid_df[covid_df.weekdays==1]
covid_df_sundays_more_cases=covid_df_sundays[covid_df_sundays.new_cases>mean_new_cases]
covid_df_sundays_more_cases
```

# GROUPING AND AGGREGATION
Lets summarise the data from day wise to month wise. We can do this by <code>.group by</code> and aggregate them using <code>.sum()</code>.



```python
covid_df_monthly=covid_df.groupby("month")[["new_cases","new_tests","new_deaths"]].sum()
covid_df_monthly
```

Lets aggregate them with <code>.mean</code>

```python
covid_df_monthly_mean=covid_df.groupby("month")[["new_cases","new_tests","new_tests"]].mean()
covid_df_monthly_mean
```

Lets make new columns which will show the cummulative sum of cases, tests and deaths as <code>total_death</code>, <code>total_cases</code>. We can use the <code>cumsum</code> method.

```python
covid_df["total_cases"]=covid_df.new_cases.cumsum()
covid_df["total_deaths"]=covid_df.new_deaths.cumsum()
covid_df["total_tests"]=covid_df.new_tests.cumsum()
covid_df
```

# MERGING DATA FROM MULTIPLE SOURCES

```python
urlretrieve('https://gist.githubusercontent.com/aakashns/8684589ef4f266116cdce023377fc9c8/raw/99ce3826b2a9d1e6d0bde7e9e559fc8b6e9ac88b/locations.csv', 
            'locations.csv')

locations_df=pd.read_csv("locations.csv")

covid_df["location"]="Italy"
```

Lets merge both the data using <code>.merge</code> method.

```python
merged_df=covid_df.merge(locations_df,on="location")
merged_df
```

```python
result_df=merged_df[["date","new_cases","new_deaths","new_tests","total_cases","total_deaths","total_tests"]]
result_df

#saving the data

result_df.to_csv("result.csv", index=None)
result_df=pd.read_csv("result.csv")
```
